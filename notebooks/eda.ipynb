# checking for multicollinearity

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in categorical_cols:
    if chi2_contingency(pd.crosstab(df['case_status'], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[categorical_cols, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

continuous_features = ['no_of_employees', 'yr_of_estab', 'prevailing_wage']

for feature in continuous_features:
    contingency_table = pd.crosstab(df['case_status'], pd.cut(df[feature], bins=10))
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    if p < 0.05:
        print(f"{feature} rejects the null hypothesis (p-value: {p:.4f})")
    else:
        print(f"{feature} fails to reject the null hypothesis (p-value: {p:.4f})")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

results = []

for feature in df.columns:
    if df[feature].dtype == 'object':
        contingency_table = pd.crosstab(df['case_status'], df[feature])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
    else:
        contingency_table = pd.crosstab(df['case_status'], pd.cut(df[feature], bins=10))
        chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    if p < 0.05:
        result = 'Rejects null hypothesis'
    else:
        result = 'Fails to reject null hypothesis'
    
    results.append({'Feature': feature, 'p-value': p, 'Result': result})

results_df = pd.DataFrame(results)
results_df
```

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor
numeric_df = df.select_dtypes(include=['int64', 'float64'])
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(numeric_df.values, i) for i in range(numeric_df.shape[1])]
vif['Feature'] = numeric_df.columns
print(vif)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming df is your DataFrame

# Calculate VIF
numeric_df = df.select_dtypes(include=['int64', 'float64'])
vif = pd.DataFrame(columns=['Feature', 'VIF'])
for i in range(numeric_df.shape[1]):
    X = numeric_df.drop(numeric_df.columns[i], axis=1)
    y = numeric_df[numeric_df.columns[i]]
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    resid = y - y_pred
    vif_value = 1 / (1 - np.var(resid) / np.var(y))
    vif = pd.concat([vif, pd.DataFrame({'Feature': [numeric_df.columns[i]], 'VIF': [vif_value]})], ignore_index=True)

print(vif)

# Calculate Tolerance
tolerance = pd.DataFrame(columns=['Feature', 'Tolerance'])
for i in range(numeric_df.shape[1]):
    X = numeric_df.drop(numeric_df.columns[i], axis=1)
    y = numeric_df[numeric_df.columns[i]]
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    r2 = 1 - (np.sum((y - y_pred) ** 2) / np.sum((y - np.mean(y)) ** 2))
    tolerance_value = 1 - r2
    tolerance = pd.concat([tolerance, pd.DataFrame({'Feature': [numeric_df.columns[i]], 'Tolerance': [tolerance_value]})], ignore_index=True)

print(tolerance)
```
